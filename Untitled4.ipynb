```python
t.drop(["day"],axis=1,inplace=True)
```

```python
x=t.drop("pm2.5",axis=1)
x
```

```python
y=t.loc[:,["pm2.5"]]
y
```

```python
ct=CT(
[
    ("OE",OE(categories=[["NW","NE","SE","cv"]]),["cbwd"]),
]
,remainder="passthrough"
)
x=pd.DataFrame(ct.fit_transform(x))
x
```

```python
xta,xte,yta,yte=TTS(x,y,random_state=2,test_size=0.2)
```

```python
pp1=PP(
[
    #("ct",ct),
    ("RFR",GSCV(RFR(criterion="poisson",n_estimators=150),param_grid={},cv=5,scoring="r2",n_jobs=-1))
],verbose=10)
pp1.fit(xta,yta)
```

```python
r2(yte,pp1.predict(xte)),pp1.named_steps["RFR"].best_score_
```

```python
pp2=PP(
[
    #("ct",ct),
    ("XG",GSCV(XGBR(learning_rate=0.1,max_depth= 12,n_estimators=500),param_grid={},cv=5,scoring="r2",n_jobs=-1))
],verbose=True)
pp2.fit(xta,yta)
```

```python
r2(yte,pp2.predict(xte)),pp2.named_steps["XG"].best_score_
```

```python
vr=VR(
[
    ("pp1",pp1),
    ("pp2",pp2),
],n_jobs=-1
,verbose=10)
vr.fit(xta,yta)
```

```python
r2(yte,vr.predict(xte))
```

```python
sn.set_theme()
sn.regplot(x=yte,y=vr.predict(xte));
mp.xlabel("Actual Value")
mp.ylabel("Predicted Value");
```

```python
sr=SR(
[
    ("pp1",pp1),
    ("pp2",pp2),
],final_estimator=LR(),passthrough=True,n_jobs=-1
,verbose=1)
sr.fit(xta,yta)
```

```python
r2(yte,sr.predict(xte))
```
